```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Load dataset
train_data = torch.tensor(pd.read_csv("train_data.csv", header=None).values, dtype=torch.float32)
train_labels = torch.tensor(pd.read_csv("train_label.csv", header=None).values.flatten(), dtype=torch.long)

test_data = torch.tensor(pd.read_csv("test_data.csv", header=None).values, dtype=torch.float32)
test_labels = torch.tensor(pd.read_csv("test_label.csv", header=None).values.flatten(), dtype=torch.long)

# Define the MLFFNN model
class MLFFNN(nn.Module):
    def __init__(self, input_size, hidden1, hidden2, output_size):
        super(MLFFNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden1)
        self.fc2 = nn.Linear(hidden1, hidden2)
        self.fc3 = nn.Linear(hidden2, output_size)
        self.activation = nn.Tanh()
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        x = self.activation(self.fc1(x))
        x = self.activation(self.fc2(x))
        x = self.fc3(x)
        return self.softmax(x)

# Model parameters
input_size = train_data.shape[1]
hidden1 = 20
hidden2 = 10
output_size = len(torch.unique(train_labels))
learning_rate = 0.002
num_epochs = 500  # Increased since pattern mode updates after each example
threshold = 4e-4  # Finer convergence criteria

# Define optimizers with fixed learning rate
optimizers = {
    "Delta Rule": lambda params: optim.SGD(params, lr=learning_rate),
    "Generalized Delta Rule": lambda params: optim.SGD(params, lr=learning_rate, momentum=0.9),  # Added momentum
    "AdaGrad": lambda params: optim.Adagrad(params, lr=learning_rate),
    "RMSProp": lambda params: optim.RMSprop(params, lr=learning_rate),
    "Adam": lambda params: optim.Adam(params, lr=learning_rate)
}

# Train and evaluate the model for each optimizer
results = {}

# Store initial weights for all optimizers to ensure fair comparison
base_model = MLFFNN(input_size, hidden1, hidden2, output_size)
initial_weights = base_model.state_dict()

for opt_name, opt_func in optimizers.items():
    # Reset model with same initial weights
    model = MLFFNN(input_size, hidden1, hidden2, output_size)
    model.load_state_dict(initial_weights)  # Ensures same start

    criterion = nn.CrossEntropyLoss()
    optimizer = opt_func(model.parameters())

    losses = []
    prev_loss = float('inf')

    for epoch in range(num_epochs):
        indices = torch.randperm(len(train_data))
        epoch_loss = 0

        for idx in indices:  # Pattern mode (update per example)
            x_sample = train_data[idx].unsqueeze(0)  # Single sample
            y_sample = train_labels[idx].unsqueeze(0)

            optimizer.zero_grad()
            output = model(x_sample)
            loss = criterion(output, y_sample)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_loss = epoch_loss / len(train_data)
        losses.append(avg_loss)

        if abs(prev_loss - avg_loss) < threshold:
            print(f"{opt_name} converged in {epoch+1} epochs.")
            break
        prev_loss = avg_loss
        print(f"{opt_name} - Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.6f}")

    results[opt_name] = {
        "epochs": epoch + 1,
        "losses": losses
    }

    # Testing the model
    with torch.no_grad():
        train_predictions = model(train_data).argmax(dim=1)
        test_predictions = model(test_data).argmax(dim=1)

        train_conf_matrix = confusion_matrix(train_labels.numpy(), train_predictions.numpy())
        test_conf_matrix = confusion_matrix(test_labels.numpy(), test_predictions.numpy())

        print(f"Confusion Matrix for {opt_name} - Training Data:")
        print(train_conf_matrix)
        print(f"Confusion Matrix for {opt_name} - Test Data:")
        print(test_conf_matrix)

        plt.figure()
        plt.plot(range(1, len(losses) + 1), losses, label=f"{opt_name}")
        plt.xlabel("Epochs")
        plt.ylabel("Average Error")
        plt.title(f"{opt_name} - Training Loss")
        plt.legend()
        plt.show()

        sns.heatmap(train_conf_matrix, annot=True, fmt='d')
        plt.title(f"{opt_name} - Training Confusion Matrix")
        plt.show()

        sns.heatmap(test_conf_matrix, annot=True, fmt='d')
        plt.title(f"{opt_name} - Test Confusion Matrix")
        plt.show()

# Compare number of epochs taken for convergence
for opt_name, result in results.items():
    print(f"{opt_name} converged in {result['epochs']} epochs.")
```
